```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading dataset
data = pd.read_csv('student_scores.csv')
data.head()
data.info()

# Plotting the data
plt.scatter(data['Hours'],data['Scores'])
```

```python
# Computing X and Y
X = data['Hours'].values
Y = data['Scores'].values
```

```python

mean_x = np.mean(X)
mean_y = np.mean(Y)

n = len(X)
```

```python
# Calculating 'm' and 'c'
num = 0
denom = 0
for i in range(n):
  num += (X[i] - mean_x) * (Y[i] - mean_y)
  denom += (X[i] - mean_x) ** 2
m = num / denom
c = mean_y - (m * mean_x)
 
# Printing coefficients
print("Coefficients")
print(m, c)
```

```python
maxx_x = np.max(X) + 10
minn_x = np.min(X) - 10
 
# line values for x and y
x = np.linspace(minn_x, maxx_x, 1000)
y = c + m * x
 
# Ploting Regression Line
plt.plot(x, y, color='#58b970', label='Regression Line')
 
# Ploting Scatter Points
plt.scatter(X, Y, c='#ef5423', label='Scatter Plot')
 
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()
plt.show()
```
